# Data Analysis

> ## Mesuring Central Tendency

- Mathematically central tendency means measuring the center or distribution of location of values of a dataset. it gives an idea of the average value of the data in the ddataset and also an indication of how widely the values are spread in the data set. That in turn helps in evaluting the chances of a new input fitting into the existing data set and hence probability of success.

There are three main measures of central tendency.
1. Mean
2. Median
3. Mode 


##### *calculating Mean and Median*

```python
import pandas as pd

# Create a Dictionary of Series 
d = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack',
    'Lee','Chanchal','Gasper','Naviya','Andres']),
    'Age':pd.Series([25,26,25,23,30,29,23,34,40,30,51,46]),
    'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65])}

# Create a DataFrame
df = pd.DataFrame(d)
df
```

```python
print("Mean Value in the Distribution")
print(df.mean(numeric_only=True))
```

```python
print("Median Value in the Distribution")
print(df.median(numeric_only=True))
```

```python
print("Mode Value in theDistribution")
print(df.mode(numeric_only=True))
```

```python
from collections import Counter

print(Counter(df['Age']))
print(Counter(df['Rating']))
```

> ## Measuring Variance

- In Statistics, variance is a measure of how far a value in data set lies from the mean value.

#### *Measuring Standard Deviation*

- Standard Deviation is square root of variance.

```python
import pandas as pd

# Create a Dictionary of Series 
d = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack',
    'Lee','Chanchal','Gasper','Naviya','Andres']),
    'Age':pd.Series([25,26,25,23,30,29,23,34,40,30,51,46]),
    'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65])}

# Create a DataFrame 
df = pd.DataFrame(d)

df.std(numeric_only=True)
```

#### *Measuring Skewness*

- It used to determine whether the data is symmetric or skewd. if the index is between -1 and 1, then the distribution is symmetric. if the index is no more than -1 then it is skewed to the left and if it is at least 1, then it is skwed to the right.

```python
skew_data = df.skew(numeric_only=True)
skew_data
```

```python
skew_data = df.skew(numeric_only=True)
skew_data
for i in skew_data:
    if i >= -1 and i <= 1:
        print("This data is a Symmetic", i)
    elif i <= -1:
        print("This data is left Skew", i)
    else:
        print("This data is Right Skew", i)
```

> ## Normal Distribution

- The Normal Distibution is a form presenting data by arranging the probability distibutionof eeach value in the data. Most values remain around the mean value making the arrangement symmetric.

- we use various function in numpy library to mathmatically calculate the values for a normal distribution. Histograms are created over which we plot the probability distribution curve.

```python
import matplotlib.pyplot as plt
import numpy as np

mu, sigma = 0.5,0.1
s = np.random.normal(mu, sigma, 100)

# Create the bins and histogram
count, bins, ignored = plt.hist(s,20)

plt.plot(bins, 1/(sigma * np.sqrt(2* np.pi)) *
         np.exp(-(bins-mu)**2 / (2* sigma**2)), linewidth=3, color='y')
plt.show()
```

> ## Chi-Square Test

Chi-Square test is a statistical method to determine if two categorical variables have a significant correlation between them. Both those variables should be from same population and they should be categorical like - yes/no, Male/Female, Red/Green etc. 

```python
from scipy import stats
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
fig, ax = plt.subplots(1,1)

linestyles = [':','--','-.','-']
deg_of_freedom = [1,4,7,6]
for df, ls in zip(deg_of_freedom,linestyles):
    ax.plot(x, stats.chi2.pdf(x,df), linestyle=ls)
    
    
plt.xlim(0,10)
plt.ylim(0,0.4)


plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Chi-Square Distribution')

plt.legend(['wavelength1','wavelength2','wavelength3', 'wavelength4'])
plt.show()
```

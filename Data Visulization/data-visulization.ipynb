##### *Formatting Line Type and Colour*

```python
import numpy as np 
import matplotlib.pyplot as plt

x = np.arange(0,10)
y = x ^ 2

# Labeling Axes and Title
plt.title("Graph Drawing")
plt.xlabel("Time")
plt.ylabel("Distance")

# Formating the Line Colour
plt.plot(x,y, 'r')

# Formating the Line Type
plt.plot(x,y, '.')
```

> ## Chart Styling

##### *Adding Annotations*

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10)
y = x ^ 2
z = x ^ 3
t = x ^ 4

#  Labeling the Axes and Title
plt.title("Drawing Table")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y)

# Annotate 
plt.annotate("Second Entry", xy=[3, 1])
plt.annotate("Third Entry", xy=[4, 6])
```

##### *Adding Legends*

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0,10)
y = x ^ 2
z = x ^ 3
t = x ^ 4

# Labeling the axes
plt.title("Graph Drawing")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y)

# Annonate
plt.annotate("Second Entry", xy=[2,0])
plt.annotate("Third Entry", xy=[4,6])

# Adding Legends
plt.plot(x,z)
plt.plot(x,t)

plt.legend(['Race1','Race2','Race3'])
```

##### *Chart Presentation Style*

```python
import numpy as np
from matplotlib import pyplot as plt

x = np.arange(0,10)
y = x ^ 2
z = x ^ 3
t = x ^ 4

# labeling the Axes and Title
plt.title("Drawing Graph")
plt.xlabel("Time")
plt.ylabel("Distance")
plt.plot(x,y)

# Annotate
plt.annotate("Second Entry", xy=[2,0])
plt.annotate("Third Entry", xy=[4,6])

# Adding Legends
plt.plot(x,z)
plt.plot(x,t)

plt.legend(['Race1','Race2','Race3'])

# Style the background
plt.style.use('fast')
plt.plot(x,z)
```

> ## Heat Map

- A heatmap contains values representing various shades of the same colour for each value to be plotted. Usually the darker shades of the chart represent higher value than lighter shade. For a very difficult value completetly colour can also used.

```python
import pandas as pd
import matplotlib.pyplot as plt

data = [{2,3,4,1},{6,3,5,2},{6,4,3,6},{9,5,6,8},{1,2,3,5}]
ind = ['I1','I2','I3','I4','I5']
cols = ['C1','C2','C3','C4']

df = pd.DataFrame(data, index=ind, columns=cols)

# Labeling
plt.title('Drawing Graph')
plt.xlabel("Index Data")
plt.ylabel("Columns Data")

plt.pcolor(df)
plt.show()
```

> ## Scatter Plots

> ## Bubble Chart

- Bubble charts display data as a cluster of circles. The required data to create bubble chart needs to have the x coordinates size the bubble of the bubbles. The colours can be supplied by the library itself.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create a Data
x = np.random.rand(40)
y = np.random.rand(40)
z = np.random.rand(40)

colors = np.random.rand(40)

# Use the Scatter Function
plt.scatter(x,y,s=z*1000,c=colors)
plt.show()
```

> ## 3D Charts

- Python is also able to creating 3d Charts. It involves adding a subplot to an existing two-dimensional plot and assigning the projection parameter as 3d.

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pylab as plt

chart = plt.figure()
chart3d = chart.add_subplot(111, projection='3d')

# Create some test Data.
x, y, z = axes3d.get_test_data(0.08)

# Plot a wireframe.
chart3d.plot_wireframe(x,y,z, color='r',rstride=15, cstride=10)

plt.show()
```

> ## Time Series

- Time Series is a series data points in which each data oint is associated withe a timestamp. A simple Example is the price of stock in the stock market at different points of time on a given day. Another example is the amount of rainfall in region at different months of the year.

```python
from datetime import datetime
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("google-stock.csv")
data
```

```python
df = pd.DataFrame(data,columns=['Date','Adj Close'])
df
```

```python
# Set the Date as Index
df['Date'] = pd.to_datetime(df['Date'])
df.index = df['Date']
del df['Date']

df.plot(figsize=(15,6))
plt.show()
```
